Pengecekan Outliers

```python
results = []

cols = data.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = data[col].quantile(0.25)
  q3 = data[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
  percent_outliers = (len(outliers) / len(data)) * 100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})


results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)


display(results_df)
```

Melakukan penangan di outliers

```python
columns_to_impute = ["Body_Temp"]

for col in columns_to_impute:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    data.loc[:, col] = data[col].clip(lower=lower_bound, upper=upper_bound)

```

##VISUALISASI DATA MATPLOTLIB & SEABORN

kita akan tammpilkan dulu dalam visualisasi data. karena datanya banyak jadi lebih relevan jika memakai bar chart.

```python
# Bar Chart (Matplotlib): Heart Rate per Umur(Age)
plt.figure(figsize=(16, 8))
data.groupby('Age')['Heart_Rate'].mean().plot(kind='bar', color='skyblue')
plt.title('Heart Rate per Umurnya')
plt.xlabel('Age')
plt.ylabel('Heart_Rate')
plt.xticks(rotation=45)
plt.show()
```

##Insight dari Visualisasi Data di Atas

Grafik menunjukkan bahwa rata-rata Heart Rate untuk setiap usia relatif stabil dan tidak banyak bervariasi.
Tidak ada pola yang jelas mengenai peningkatan atau penurunan Heart Rate seiring bertambahnya usia.
Masalah pada Visualisasi

Terlalu banyak kategori usia pada sumbu X, sehingga label usia menjadi sulit dibaca dan membingungkan.
